```python
import pandas as pd
import numpy as np

data = pd.read_csv("steam_reviews.csv")
# data = data[:200000]
data
```

```python
reviews = data["review"]
AllWords = []
UniqueWords = []
UsefulWords = []
```

```python
for review in reviews:
    review = str(review)  # Apparently some of the reviews get classified as floats?
    words = review.split()
    for word in words:
        AllWords.append(word)
```

```python
# UniqueWords = list(set(AllWords)) # Remove duplicates from AllWords to get UniqueWords
seen = set()
seen_add = seen.add
UniqueWords = [x for x in AllWords if not (x in seen or seen_add(x))]
```

```python
# Placeholders for real work
UsefulWords = UniqueWords[1 : len(UniqueWords) // 200]
CleanedReviews = reviews
```

```python
df = pd.DataFrame(
    np.zeros([data.shape[0], len(UsefulWords)], dtype=np.int8), columns=UsefulWords
)  # Add 5 rows for funny, helpful, hour_played, is_early_access_review, and recommendation.
```

```python
for idx, review in enumerate(CleanedReviews):
    for word in review.split():
        df.iloc[idx][UniqueWords.index(word)] += 1
    print("idx = " + str(idx))
    if idx > 10:
        break
```
